# Visualizing Your Data

While being able to interpret and explain what's in a data set is a key skill for any data scientist or analyst, being able to communicate your findings is just as important. 

There are two very commonly used packages for visualization with Python:
- matplotlib
- seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg')
cars
```

## Bar Graphs
Used to determine the quantity of categorical or discrete features. 

```python

frequency_origin = cars['origin'].value_counts()
frequency_origin
```

```python
fig, ax = plt.subplots() # two objects, object fig and object ax

ax.bar(frequency_origin.index, frequency_origin) # object ax, frequency_origin.index will give you the names for x-axis, and frequency_origin will give you the numbers in the y axis.
```

```python
# for model year:
model_year = cars['model_year'].value_counts()

fig, ax = plt.subplots()

ax.bar(model_year.index, model_year)
```

## Histograms
Measure the distribution of values for one quantitative values, divided into separate bins. 
different from bar graph because x axis is also a number not a name

```python
#calling subplot object again
fig, ax = plt.subplots()
ax.hist(cars['displacement'], bins = 15); 
   # to remove the text below, just add ; to the end of this line
   # bin number means how many groupings we want
```

## Box Plots
Measures the distribution of a quantitative array of values by visually describing the summary statistics: 
- min
- max
- median
- 25th and 75th percentile

```python
cars['displacement'].describe()
```

```python
cars['displacement'].describe()
fig, ax = plt.subplots()
ax.boxplot(cars['displacement'], vert=False);   

# vert=False will rotate the boxplot 90 degrees
# the box is the interquartile range, one edge is the 25 percentile, the other is the 75 percentile
# whiskers at the end are min and max 68 and 455, but we can still have outliers because the min max are not true min max, they're calculated by using the interquartile range
#   times 1.5
# lower quartile is the median of the lower region separated by the median; upper quartile is the median of the upper region
```

## Scatterplots
The values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.

```python
fig, ax = plt.subplots()
ax.scatter(x = cars['displacement'], y = cars['mpg']);
```

## Matplotlib Parameters


```python
fig, ax = plt.subplots()
fig.set_size_inches(6,6)
ax.set_title("Scatterplot for Cars", fontdict={'fontsize':26, 'color':'orange'})
ax.set_facecolor('black')
ax.scatter(x = cars['displacement'], y = cars['mpg']);
```

## Seaborn


```python

# make customization more straightforward
plt.figure(figsize=(6,6));
plt.title('Seaborn Title', fontdict = {'fontsize':26})
sns.scatterplot(x='displacement', y='mpg', data=cars, size = 'cylinders', hue = 'origin'); # size and hue are other aspects of data we can display making it a dynamic chart!
# this will be the last line. customization will be before this line.
```
