#Intro to Exploratory Data Analysis

```python
  #imports
import pandas as pd
import seaborn as sns
```

```python
#read in the dataset
autos = sns.load_dataset('mpg')
autos
```

```python
#output the first five rows
autos.head()
```

```python
#see some details
autos.columns
```

```python
autos.info()
```

```python
autos.shape # rows first columns second
```

```python
#data types
autos.dtypes
```

```python
#check for missing values
autos.isnull()
```

```python
autos.isnull().sum()
```

```python
#sort values
autos.sort_values(by='horsepower', ascending = False)
```

```python
#fill nulls
autos.fillna(1000).sort_values(by='horsepower')
#drop nulls
autos.dropna()
#reassign to object
autos = autos.dropna()
```

```python
autos
```

```python
#check for duplicates
autos.duplicated().sum()
```

```python
#check summary statistics
autos.describe()
```

```python
#check summary statistics for just one column
autos['mpg']
```

```python
autos['mpg'].describe()
```

```python
#histogram
sns.histplot(autos['mpg']) # specify the column of data you wanna show, and it'll show the frequency
```

```python
#scatterplot
sns.scatterplot(x='displacement', y='mpg', data=autos)
```

```python
#pairplot for every single possible relationship between two variables
sns.pairplot(autos)
```

```python
#dataframe subset 
#requires double brackets because it has to display a list
autos_subset = autos[['mpg', 'displacement', 'horsepower']]
autos_subset
```

```python
#save to directory on Google Drive, then you can download to your local computer
autos_subset.to_csv("autos_subset.csv")
```
