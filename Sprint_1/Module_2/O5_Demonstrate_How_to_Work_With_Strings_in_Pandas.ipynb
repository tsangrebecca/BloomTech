
method	description

['column'].str.lower()	converts characters to lower case

['column'].str.upper()	converts characters to upper case

['column'].str.title() just cap the first letter of a word

['column'].str.len()	returns the length of each item

['column'].str.strip()	removes white space

['column'].str.split('_')	separate on the given character

df.dropna(subset=['Value'], inplace=True) -- To drop the rows with missing values in the 'Value' column and specify the column to consider using the subset parameter as a list. The inplace=True parameter ensures that the changes are made directly in the original DataFrame.

```python
# capitalize city names to title case
ufo_df['city'] = ufo_df['city'].str.title()

# convert state and country abbreviations to all upper case
ufo_df['state'] = ufo_df['state'].str.upper()
ufo_df['country'] = ufo_df['country'].str.upper()

# display the correct dataframe
display(ufo_df.head())
```

```python
ufo_df.dropna(subset=['country'], inplace=True)
ufo_df.isnull().sum()
```
