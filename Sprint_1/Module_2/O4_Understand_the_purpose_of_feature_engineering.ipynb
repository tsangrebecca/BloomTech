```python
import pandas as pd

mms = pd.read_csv('https://tinyurl.com/mms-statistics')

display(mms.head())
```

```python
import math

pie = math.pi
print(pie)
```

```python
# add volume hence density columns to the data
# diameter is in mm, convert to cm

# create volume column
mms['volume_in_cm^3'] = (4/3)*(pie)*(mms['diameter']*0.1/2)**3

# Create density column
mms['density_in_g/cm^3'] = mms['mass']/mms['volume_in_cm^3']

# see the new columns
display(mms.head())
```

```python
# create radius column
mms['radius'] = mms['diameter']/2
mms.head()
```
