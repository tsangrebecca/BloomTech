```python
import pandas as pd
import seaborn as sns

crashes = sns.load_dataset("car_crashes")

mycols = ['total', 'speeding', 'alcohol']

# only load the columns of interest
crashes = crashes[mycols]
crashes.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
import statsmodels.api as sm

Y = crashes['total']
X = crashes['alcohol']

# create model with X, Y data
X = sm.add_constant(X)
model = sm.OLS(Y, X)

# fit the model
results = model.fit()

# interpret results
print(results.t_test([1,0]))
```

```python
# coef is the slope
# std err
# t is t-value
# P>|t| is the p-value

# remember t-stat = sample mean - null hypo mean / SE
# adusted t-stat = sample slope - null hypo slope / SE
# so t = 5.8578 - 0 / 0.921 = 6.36 which matches the t-value as well

# p-value is essentially zero, so we can reject the null hypothesis.
```
